```python
# Import the required libraries & modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

```python
gbc_patients = pd.read_excel("GBC.xlsx")
```

```python
filtered_data = gbc_patients.dropna(subset=[ 'Status','Date_event', 'Date_diagnosis'])
filtered_data.shape
```

```python
# Calculate the difference of Date_event and Date_diagnosis in months
filtered_data['Time_overall_study'] = ((filtered_data['Date_event'] - filtered_data['Date_diagnosis']) / np.timedelta64(1, 'M')).astype(float)

filtered_data.head()
```

```python
#Time_overall_study feature
value_to_delete = -8.01659171646235

# Delete the row with the specified value
filtered_data = filtered_data[filtered_data['Time_overall_study'] != value_to_delete]

filtered_data.shape
```

```python
# Mapping of values to be replaced
status_mapping = {'Dead': 1, 'No': 0}

# Replace values in the "Status" column
filtered_data['Status'] = filtered_data['Status'].replace(status_mapping)
```

```python
# List of columns to drop
columns_to_drop = ['MR_NO', 'COUNTRY','STATE_NAME', 'PINCODE', 'CITY_NAME','First_Visit_Date', 'Date_diagnosis','Date_event','UREA_from_UnC_test','CREATININE_from_UnC_test','ESTIMATED AVERAGE GLUCOSE (eAG)','HbA1c', 'CREATININE_from_CREATININE_test','Site_metastasis', 'TNM_classification','Family_cancer_history','Tumour_grade_final', 'Pathology', 'Metastasis_final','Mets_site_final','Treatment_clean', 'Treatment', 'Cholecystectomy_TMC', 'Outside_cholecystectomy','Diabetes','Cancer_presentation_final','Thyroidism','HTN']

# Drop the specified columns
df = df.drop(columns=columns_to_drop)
```

```python
df.columns
```

```python
# Drop rows with any missing values
clean_data = df.dropna()

print("Cleaned DataFrame:")
print(clean_data.shape)
```

```python
# Select categorical columns
categorical_columns = clean_data.select_dtypes(include=['object'])

# Select continuous columns (excluding categorical)
continuous_columns = clean_data.select_dtypes(exclude=['object'])
```

```python
# Calculate correlation matrix
correlation_matrix = continuous_columns.corr()

# Set up the matplotlib figure
plt.figure(figsize=(12, 10))

# Generate a heatmap
sns.set(font_scale=1)  # Adjust font size for annotations
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", center=0, linewidths=0.5)

# Add plot title
plt.title("Correlation Heatmap")

# Show the plot
plt.tight_layout()
plt.show()

# Function to remove highly correlated variables
def remove_highly_correlated(continuous_columns, threshold=0.9):
    corr_matrix = continuous_columns.corr().abs()
    upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
    to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > threshold)]
    continuous_columns = continuous_columns.drop(columns=to_drop)
    return continuous_columns

# Remove highly correlated variables
threshold = 0.9
continuous_columns_no_high_correlation = remove_highly_correlated(continuous_columns, threshold)

print("Original DataFrame shape:", continuous_columns.shape)
print("DataFrame after removing highly correlated variables shape:", continuous_columns_no_high_correlation.shape)
```

```python
##performing one hot encoding to the categorical variables in order to pass the weightage of categorical variables in the form of numbers to our model. 
```

```python
# Perform one-hot encoding on categorical columns
encoded_categorical = pd.get_dummies(categorical_columns, drop_first=True)
```

```python
print(categorical_columns.isnull().sum())
print("*********************************")
print(continuous_columns_no_high_correlation.isnull().sum())
```

```python
# Concatenate the encoded categorical columns with continuous columns
working_dataframe = pd.concat([encoded_categorical, continuous_columns_no_high_correlation], axis=1)
```

```python
df = working_dataframe.copy()
```

```python
from lifelines import CoxPHFitter 

cph = CoxPHFitter(alpha=0.05)
cph.fit(df, duration_col='Time_overall_study', 
              event_col='Status',
              show_progress = True)
```
